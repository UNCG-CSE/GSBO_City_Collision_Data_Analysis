```python
import time
from time import mktime
from datetime import datetime
#2018-12-19 00:00:00
def getTimeCat(Datetime):
    # extract time categories
    
    ts = Datetime.time()
     

    #  Morning = 6am to 8pm
    mornStart = datetime.time(6, 0, 1)
    mornEnd = datetime.time(20, 0, 0)

    #  Night = 8pm to 6am
    midStart = datetime.time(20, 0, 1)
    midEnd = datetime.time(6, 0, 0)

    if is_time_between(mornStart, mornEnd, ts):
        return "d"
    elif is_time_between(midStart, midEnd, ts):
        return "n"
    return "null"
```

```python
from datetime import datetime, time

def is_time_between(begin_time, end_time, check_time=None):
    # If check time is not given, default to current UTC time
    check_time = check_time or datetime.utcnow().time()
    if begin_time < end_time:
        return check_time >= begin_time and check_time <= end_time
    else: # crosses midnight
        return check_time >= begin_time or check_time <= end_time
```

```python
import datetime
i= 0
morning_dates = []
total_day_accidents = 0

night_dates = []
total_night_accidents = 0
for index,date in data_accidents_list.iterrows():
        if isinstance(date['ACCIDENT_DATE'], datetime.date):
            a = getTimeCat(date['ACCIDENT_DATE'])
            if a == "d":
                total_day_accidents = total_day_accidents + 1
                morning_dates.append(date['ACCIDENT_DATE'].date())
            elif a == "n":
                total_night_accidents = total_night_accidents + 1
                night_dates.append(date['ACCIDENT_DATE'].date())

unique_morning_dates = len(set(morning_dates))
unique_night_dates =  len(set(night_dates))


print(total_day_accidents)
print(unique_morning_dates)

print("--")

print(total_night_accidents)
print(unique_night_dates)
                
                
morning_accident_lamda = total_day_accidents/unique_morning_dates
night_accident_lamda = total_night_accidents/unique_night_dates
print(morning_accident_lamda)
print(night_accident_lamda)
```

```python
print(morning_accident_lamda)
import math
x= []
y = []
for  k in range(0,100):
    y_value= ((morning_accident_lamda**k)*(math.exp(-(morning_accident_lamda))))/math.factorial(k)
    x_value = k
    x.append(x_value)
    y.append(y_value)
y_final_value = []
for element in y:
    y_final_value.append((element - min(y))/(max(y)))
plt.plot(x,y_final_value)
#prob of these accidents in this time
```
